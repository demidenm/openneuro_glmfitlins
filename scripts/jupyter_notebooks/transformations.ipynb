# Cumulative Datasets and Subjects on OpenNeuro


````python
# below code is a modification of Claude req/ssl from git repos maintained by Joe Wexler.

# OpenNeuro data
ssl_context = ssl._create_unverified_context()
url = 'https://raw.githubusercontent.com/jbwexler/openneuro_metadata/main/metadata.csv'
with urllib.request.urlopen(url, context=ssl_context) as response:
    df = pd.read_csv(response)

df['made_public'] = pd.to_datetime(df['made_public'])
df = df.dropna(subset=['made_public'])
df['yr_mon'] = df['made_public'].dt.to_period('M')

# 1. cumulative datasets
datasets_cumulative = df.groupby('yr_mon').size().cumsum()

# 2. cumulative subjects
subjects_cumulative = df.groupby('yr_mon')['num_subjects'].sum().cumsum()

# 3. Fetch derivative repositories from OpenNeuroDerivatives
def fetch_derivative_repos():
    """Fetch all derivative repositories from OpenNeuroDerivatives organization"""
    repos = []
    page = 1
    per_page = 100
    
    while True:
        url = f"https://api.github.com/orgs/OpenNeuroDerivatives/repos?page={page}&per_page={per_page}"
        
        try:
            response = requests.get(url)
            response.raise_for_status()
            data = response.json()
            
            if not data:  # No more repositories
                break
                
            repos.extend(data)
            page += 1
            
        except requests.exceptions.RequestException as e:
            print(f"Error fetching repositories: {e}")
            break
    
    return repos

# Fetch and process derivative repository data
print("Fetching derivative repositories...")
all_repos = fetch_derivative_repos()

# Filter for fMRIPrep and MRIQC repositories
fmriprep_repos = [repo for repo in all_repos if 'fmriprep' in repo['name'].lower()]
mriqc_repos = [repo for repo in all_repos if 'mriqc' in repo['name'].lower()]

# Process fMRIPrep repositories
if fmriprep_repos:
    fmriprep_data = []
    for repo in fmriprep_repos:
        fmriprep_data.append({
            'name': repo['name'],
            'created_at': repo['created_at'],
            'updated_at': repo['updated_at']
        })
    
    fmriprep_df = pd.DataFrame(fmriprep_data)
    fmriprep_df['created_at'] = pd.to_datetime(fmriprep_df['created_at'])
    fmriprep_df['yr_mon'] = fmriprep_df['created_at'].dt.to_period('M')
    fmriprep_cumulative = fmriprep_df.groupby('yr_mon').size().cumsum()
    
    print(f"Found {len(fmriprep_repos)} fMRIPrep repositories")
else:
    print("No fMRIPrep repositories found")
    fmriprep_cumulative = pd.Series(dtype='int64')

# Process MRIQC repositories
if mriqc_repos:
    mriqc_data = []
    for repo in mriqc_repos:
        mriqc_data.append({
            'name': repo['name'],
            'created_at': repo['created_at'],
            'updated_at': repo['updated_at']
        })
    
    mriqc_df = pd.DataFrame(mriqc_data)
    mriqc_df['created_at'] = pd.to_datetime(mriqc_df['created_at'])
    mriqc_df['yr_mon'] = mriqc_df['created_at'].dt.to_period('M')
    mriqc_cumulative = mriqc_df.groupby('yr_mon').size().cumsum()
    
    print(f"Found {len(mriqc_repos)} MRIQC repositories")
else:
    print("No MRIQC repositories found")
    mriqc_cumulative = pd.Series(dtype='int64')

# Create the plot with 2 subplots
fig, axes = plt.subplots(2, 1, figsize=(10, 8), sharex=True)

# Fig 1: datasets (including fMRIPrep and MRIQC repos)
axes[0].plot(datasets_cumulative.index.to_timestamp(), datasets_cumulative.values, marker='o', label='OpenNeuro Datasets')
if not fmriprep_cumulative.empty:
    axes[0].plot(fmriprep_cumulative.index.to_timestamp(), fmriprep_cumulative.values, marker='s', color='green', label='fMRIPrep Repos')
if not mriqc_cumulative.empty:
    axes[0].plot(mriqc_cumulative.index.to_timestamp(), mriqc_cumulative.values, marker='^', color='red', label='MRIQC Repos')
axes[0].set_ylabel('# of Datasets')
axes[0].set_title('Datasets Since 2016')
axes[0].grid(True)
axes[0].legend()

# Fig 2: subjects
axes[1].plot(subjects_cumulative.index.to_timestamp(), subjects_cumulative.values, marker='o', color='orange')
axes[1].set_ylabel('# of Subjects')
axes[1].set_title('Subjects Since 2016')
axes[1].set_xlabel('Date')
axes[1].grid(True)

today = datetime.now().strftime('%Y-%m-%d')

footnote_text = f"Last updated: {today}"
axes[1].text(0.02, -0.15, footnote_text, transform=axes[1].transAxes, 
             fontsize=8, style='italic', color='gray')

plt.tight_layout()
plt.show()
````

# Folder Size and File Counts

````python
def extract_stats_from_readme(file_path):
    """Extract size, number of files, and number of subjects from README.md"""
    with open(file_path, 'r') as f:
        content = f.read()
    
    # Pattern looks for strings like "The size of the Fitlins Derivatives for ds000001 balloonanalogrisktask is 13G with 8816 files."
    pattern = r"size of .+ for (\w+) (\w+) is (\d+\.?\d*)(\w) with (\d+) files"
    match = re.search(pattern, content, re.IGNORECASE)
    
    # Pattern for finding number of subjects
    subject_pattern = r"task for (\d+) subjects"
    subject_match = re.search(subject_pattern, content, re.IGNORECASE)
    
    num_subjects = None
    if subject_match:
        num_subjects = int(subject_match.group(1))
    
    if match:
        study_id = match.group(1)
        task_name = match.group(2)
        size_value = float(match.group(3))
        size_unit = match.group(4)
        num_files = int(match.group(5))
        
        # Convert size to GB for consistent comparison
        if size_unit.lower() == 'g':
            size_gb = size_value
        elif size_unit.lower() == 'm':
            size_gb = size_value / 1024
        elif size_unit.lower() == 'k':
            size_gb = size_value / (1024 * 1024)
        else:
            size_gb = size_value / (1024 * 1024 * 1024)
            
        return {
            'study_id': study_id,
            'task_name': task_name,
            'size_gb': size_gb,
            'size_display': f"{size_value}{size_unit}",
            'num_files': num_files,
            'num_subjects': num_subjects
        }
    return None

def find_readme_files(root_dir):
    """Find all README.md files in group_* subfolders"""
    results = []
    
    for study_dir in os.listdir(root_dir):
        study_path = os.path.join(root_dir, study_dir)
        
        if os.path.isdir(study_path):
            for subdir in os.listdir(study_path):
                if subdir.startswith('group_'):
                    subdir_path = os.path.join(study_path, subdir)
                    readme_path = os.path.join(subdir_path, 'README.md')
                    
                    if os.path.exists(readme_path):
                        stats = extract_stats_from_readme(readme_path)
                        if stats:
                            results.append(stats)
    
    return results
````

## Focus on Subject-Level Models

Modifying multiple files can be tedious, making programmatic transformations more efficient. While group- and participant-level modifications are often simpler (e.g., relabeling groups by sex or condition), this notebook focuses on subject-level transformations which benefit the most from automation. 

In this example, we are solely focused on using the general linear model (GLM) for analyzing time-series data at the subject level, which is given by:

$$
Y = X\beta + \varepsilon
$$

where:
- **Y** is the observed time-series data (e.g., BOLD response in fMRI),
- **X** is the design matrix containing predictors (e.g., task regressors, motion parameters),
- **** represents the regression coefficients (parameter estimates),
- **** is the residual error term, assumed to be normally distributed: \( \varepsilon \sim \mathcal{N}(0, \sigma^2 I) \).

## PyBIDS Transformations Summary

[Tals PyBIDS Transformations Document](https://docs.google.com/document/d/1uxN6vPWbC7ciAx2XWtT5Y-lBrdckZKpPdNUNpwRxHoU/edit?tab=t.0) covers several transformation options. Within the context of BIDS Stats-Models, using "pybids-transforms-v1", here's a summary of the transformation names and their functions:

**Core Transformations: In order of presentation**
* **Factor:** Converts categorical variables into dummy-coded (binary) indicator variables.
* **Convolve:** Convolves time-series data with a hemodynamic response function (HRF) model.
* **Filter:** Subsets rows based on a boolean expression.
* **And/Or:** Performs logical AND or OR operations on multiple columns, producing a single boolean output.
* **Assign:** Copies or reassigns values from one column to another, potentially changing attributes like onset or duration.
* **Demean:** Centers a column by subtracting its mean.
* **DropNA:** Removes rows containing missing values.
* **Group:** Creates a named group of variables for easier referencing.
* **Lag:** Shifts a column's values by a specified number of time points.
* **Log10:** Computes the base-10 logarithm of a column.
* **Not:** Performs logical negation on a column.
* **Orthogonalize:** Removes the linear relationship between columns.
* **Product:** Calculates the row-wise product of multiple columns.
* **Rename:** Changes the name of a column.
* **Replace:** Substitutes values within a column.
* **Resample:** Changes the sampling rate of a column.
* **Scale:** Standardizes a column by centering and scaling.
* **Select:** Retains only specified columns, removing others.
* **Split:** Splits a column into multiple columns based on the values of other columns.
* **Sum:** Calculates the row-wise sum of multiple columns.
* **Threshold:** Applies a cutoff to a column's values, optionally binarizing them.

***Not Reviewed***

* **Copy:** Creates a duplicate of a column under a new name.
* **Delete:** Removes specified columns from the dataset.
* **ToDense:** Converts sparse (onset/duration) data to dense (continuous) time-series data.

As described in the [BIDS Extension Proposal 2](https://docs.google.com/document/d/1bq5eNDHTb6Nkx3WUiOBgKvLNnaa5OMcGtD0AZ9yms2M/edit?tab=t.0), the nodes are performed in the order of:
* run  subject  dataset  
* run  session  subject  dataset  
* run  subject  
* subject  dataset

### Assign Duration / RT Regressor

While the events files have duration columns, `Assign` transformations can also used to modify how the value, ampltiude, onset or duration is modified. For example, if you are wanting to modify the amplitude by some value (ratings) or duration be RT, you can use this to modify those values. Often, for each column in df this should be added, as the list may omit some modifications.
```python
{
    "Name": "Assign",
    "Input": ["modulating_columnname"], # required column that has values that will be applied to a target
    "Target": ["target_columnname"], # required, what column we are wanting to modify values, onset, duration or amplitude for
    "TargetAttr": "<type>", # string or list of values (e.g. "amplitude" or ["amplitude","onset"]) to assign the input values for target colum
    "Output": ["gain_demean_modulated", "loss_demean_modulated"] # option list of values to same out the columns
}

```

In this example, we will set the RT duration. First, we willl use `Copy` to make a copy of trial_type to rt_regressor and rename all trial types to `rt` (note, we modified events so 'NoResp' was renamed to 'noresp' as PyBIDS automatically converts 'NoResp' to 'no_resp' *which does not exist in the data*). There are alternate ways of using `Threshold` as was done [here](https://github.com/bids-standard/model-zoo/blob/main/model-zoo/exhibits/narps/model-narps_smdl.json) but this may unknowningly assign values if unexpected differences exist in the data that may modify the values.

````python
# Assuming design_events has columns: 'onset', 'duration', 'amplitude', and 'trial_type'
trial_types = design_events['trial_type'].unique()
convolved_results = {}

for trial_type in trial_types:
    # Filter the design matrix for this trial type
    events_for_type = design_events[design_events['trial_type'] == trial_type]
    
    # Extract the numeric columns and ensure they're float type
    onsets = events_for_type['onset'].values.astype(float)
    durations = events_for_type['duration'].values.astype(float)
    amplitudes = events_for_type['amplitude'].values.astype(float)
    
    # Skip if there are no events for this type
    if len(onsets) == 0:
        continue
    
    # Perform the convolution for this trial type
    out_convolved, names = compute_regressor(
        [onsets, durations, amplitudes],  # Pass as a list of arrays
        hrf_model='spm',
        frame_times=frame_times,
        oversampling=50
    )
    
    # Store the result
    convolved_results[trial_type] = out_convolved
    
design_matrix = np.zeros((len(frame_times), len(trial_types)))

# Fill in each column with the corresponding convolved data
for i, trial_type in enumerate(trial_types):
    if trial_type in convolved_results:
        design_matrix[:, i] = convolved_results[trial_type].flatten()

# Create a DataFrame if needed
design_df = pd.DataFrame(design_matrix, columns=trial_types)
````
